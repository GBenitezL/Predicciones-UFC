```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("data.csv")
```

```python
df.head()
```

```python
df.describe()
```

```python
plt.hist(df['B_age'])
plt.show()
```

```python
plt.pie(df.groupby(['Winner'])['Winner'].count(), labels = ['Blue', 'Draw', 'Red'], colors = ['b', 'g', 'r'], explode = [0, 0.2, 0])
plt.title("Peleas Ganadas")
plt.show()
```

```python
plt.hist(df['B_total_time_fought(seconds)'])
plt.title("Tiempo de peleas (Segundos)")
plt.show()
```
